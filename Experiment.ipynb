# plot code test

```python
import matplotlib.pyplot as plt

p1 = Alice
p2 = Charlie

def plot_ratings(*args):

    players = args
    
    fig, ax = plt.subplots()
    
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
    
    for player in players:
        
        game = range(1, len(player.rating_history)+1)
        mu = [rating.mu for rating in player.rating_history]
        sigma = [rating.sigma for rating in player.rating_history]
        ci_lower = [a - b for a, b in zip(mu, sigma)]
        ci_upper = [a + b for a, b in zip(mu, sigma)]
        
        color = colors[players.index(player)]
        ax.plot(game, mu, color=color, label=player.name)
        ax.fill_between(game, ci_lower, ci_upper, alpha=0.1, color=color)
        ax.plot(game, mu, 'o', markersize=3, color=color)
    
    ax.legend()
    ax.set_xlabel('Number of games (N)')
    ax.set_ylabel('Trueskill rating')
    
    return
    
plot_ratings(Alice, Charlie)
```

# 3-way game

```python
from RepeatGames import play_hex
from HexBoard import Agent, HexBoard

N = 50
game = 2
boardsize = 3

Alice = Agent(name='random', searchby='random')
Bob = Agent(name='ab3', searchby='alphabeta', depth=3)
Charlie = Agent(name='ab4', searchby='alphabeta', depth=4)

print('Initial rating:')
print("Rating Alice:", Alice.rating)
print("Rating Bob:", Bob.rating)
print("Rating Charlie:", Charlie.rating)

for i in range(0, N):
    
    play_hex(game, Alice, Bob, boardsize, show_midgame=False, show_endgame=False, analysis=False)
    play_hex(game, Alice, Charlie, boardsize, show_midgame=False, show_endgame=False, analysis=False)
    play_hex(game, Bob, Charlie, boardsize, show_midgame=False, show_endgame=False, analysis=False)
    
print('Final rating')
print("Rating Alice:", Alice.rating)
print("Rating Bob:", Bob.rating)
print("Rating Charlie:", Charlie.rating)

Alice.plot_rating_history(Bob, Charlie)
```
