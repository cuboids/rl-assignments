```python
from RepeatGames import play_hex
from HexBoard import Agent, HexBoard

game = 2
boardsize = 7
M = [100,300,500,900]
for  j in M:
    print("Given N = {N:}".format(N = j))
    Alice = Agent(name='mcts N{N:} Cp0'.format(N = j), searchby='mcts', hyperpars={'N': j, 'Cp': 0, 'timelimit': 2})
    Bob = Agent(name='mcts N{N:} Cp0.01'.format(N = j), searchby='mcts', hyperpars={'N': j, 'Cp': 0.01, 'timelimit': 2})
    Charlie = Agent(name='mcts N{N:} Cp1'.format(N = j), searchby='mcts', hyperpars={'N': j, 'Cp': 1, 'timelimit': 2})
    Dylan = Agent(name='mcts N{N:} Cp5'.format(N = j), searchby='mcts', hyperpars={'N': j, 'Cp': 5, 'timelimit': 2})
    
    print('Initial rating:')
    print("Rating Alice:", Alice.rating)
    print("Rating Bob:", Bob.rating)
    print("Rating Charlie:", Charlie.rating)
    print("Rating Dylan:", Dylan.rating)


    for i in range(0, 5):
    
        play_hex(game, Alice, Bob, boardsize, show_midgame=False, show_endgame=False, analysis=False,seed = 1e7 * i)
        play_hex(game, Charlie, Alice, boardsize, show_midgame=False, show_endgame=False, analysis=False,seed = 1e7 * i)
        play_hex(game, Bob, Charlie, boardsize, show_midgame=False, show_endgame=False, analysis=False,seed = 1e7 * i)
        play_hex(game, Alice, Dylan, boardsize, show_midgame=False, show_endgame=False, analysis=False,seed = 1e7 * i)
        play_hex(game, Charlie, Dylan, boardsize, show_midgame=False, show_endgame=False, analysis=False,seed = 1e7 * i)
        play_hex(game, Bob, Dylan, boardsize, show_midgame=False, show_endgame=False, analysis=False,seed = 1e7 * i)
        
    print('Final rating',i+1,"th times")
    print("Rating Alice:", Alice.rating, "parameter",Alice.name)
    print("Rating Bob:", Bob.rating, "parameter", Bob.name)
    print("Rating Charlie:", Charlie.rating, "parameter", Charlie.name)
    print("Rating Dylan:", Dylan.rating, "parameter", Dylan.name)
    
    Alice.plot_rating_history(Bob, Charlie, Dylan)
```

```python
from RepeatGames import play_hex
from HexBoard import Agent, HexBoard

game = 20
boardsize = 5

iteration = [100,200,300,400,500,600]
iteration_2 = [100,200,300,400,500,600]

for N in range(len(iteration)):
    item1 = iteration[N]
    
    iteration_2.remove(item1)
    
    if len(iteration_2) == 0:
        break
        
    for item2 in iteration_2:
        Alice = Agent(name='mcts N{N:} '.format(N = item1), searchby='mcts', hyperpars={'N': item1, 'Cp': 1, 'timelimit': 2})
        Bob = Agent(name='mcts N{N:}'.format(N = item2), searchby='mcts', hyperpars={'N': item2, 'Cp': 1, 'timelimit': 2})

        print('Initial rating:')
        print("Rating Alice:", Alice.rating)
        print("Rating Bob:", Bob.rating)






        play_hex(game, Alice, Bob, boardsize, show_midgame=False, show_endgame=False, analysis=False)

        print('Final rating')
        print("Rating Alice:", Alice.rating, "parameter",Alice.name)
        print("Rating Bob:", Bob.rating, "parameter", Bob.name)


        print()

        Alice.plot_rating_history(Bob)
```

```python
from RepeatGames import play_hex
from HexBoard import Agent, HexBoard

Alice = Agent(name='Alice', searchby='alphabeta')
Bob = Agent(name='Bob', searchby='mcts', hyperpars={'N': 200, 'Cp': 1, 'timelimit': 1})

print('Initial rating:')
print("Rating Alice:", Alice.rating)
print("Rating Bob:", Bob.rating)

result = play_hex(100, Alice, Bob, 5, show_midgame=False, show_endgame=False)

print('Final rating')
print("Rating Alice:", Alice.rating)
print("Rating Bob:", Bob.rating)

winners = [v['winner'] for k,v in result['games'].items()]
print(f'Alice win:{winners.count("Alice")} Bob win:{winners.count("Bob")}')
Alice.plot_rating_history(Bob)
```
